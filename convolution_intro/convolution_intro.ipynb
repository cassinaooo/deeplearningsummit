```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import scipy.ndimage.filters as filter
```

```python
def normalize(matrix):
    sum = np.sum(matrix)
    if sum > 0.:
        return matrix / sum
    else:
        return matrix
```

```python
kernel = np.array([[-1.,-1.,-1.],
                    [-1.,8.,-1.],
                    [-1.,-1.,-1.]])

#kernel = np.array([[0.,1.,0.],
#                    [1.,-4.,1.],
#                    [0.,1.,0.]])
```

```python
image=mpimg.imread('Lenna.png')[:,:-1,0]
shape = image.shape
newimage = np.ndarray(shape,dtype=np.float)
poorimage = np.ndarray(shape,dtype=np.float)
supershape = (shape[0] + 2,shape[1] + 2) 
supermatrix = np.zeros(supershape,dtype=np.float)
supermatrix[1:-1,1:-1] = image
kernel
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(16, 8))
ax1.imshow(image,cmap=plt.cm.gray)
ax1.axis('off')
kernel = normalize(kernel)
newimage = filter.convolve(image,kernel,mode='constant', cval=0)
ax2.imshow(newimage,cmap=plt.cm.gray)
ax2.axis('off')
plt.tight_layout()
```
